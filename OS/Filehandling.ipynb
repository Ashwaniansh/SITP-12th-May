File Handling in Python

File handling means working with files like:

Creating files
Reading files
Writing data into files
Updating files
Deleting files

Python provides built-in functions for file handling.

1. Opening a File

Syntax:

file = open("filename.txt", "mode")

Example:

file = open("data.txt", "r")
2. File Modes
Mode	Meaning
"r"	Read file
"w"	Write file (old data removed)
"a"	Append data
"x"	Create new file
"rb"	Read binary file
"wb"	Write binary file
3. Reading a File
Example 1: Read Full File
file = open("data.txt", "r")

content = file.read()

print(content)

file.close()
Example 2: Read Line by Line
file = open("data.txt", "r")

for line in file:
    print(line)

file.close()
4. Writing into a File
file = open("data.txt", "w")

file.write("Hello Python")

file.close()

Output in file:

Hello Python
5. Appending Data
file = open("data.txt", "a")

file.write("\nWelcome")

file.close()

Now file contains:

Hello Python
Welcome
6. Using with open() (Best Method)

No need to close file manually.

with open("data.txt", "r") as file:
    print(file.read())
7. Create a New File
file = open("newfile.txt", "x")

file.close()
8. Delete a File
import os

os.remove("data.txt")
9. Check File Exists or Not
import os

if os.path.exists("data.txt"):
    print("File exists")
else:
    print("File not found")
Important Functions
Function	Work
read()	Read complete file
readline()	Read one line
readlines()	Read all lines in list
write()	Write data
close()	Close file
Simple Example Program
with open("student.txt", "w") as file:
    file.write("Ansh\n")
    file.write("MCA")

with open("student.txt", "r") as file:
    print(file.read())

Output:

Ansh
MCA
Advantages of File Handling
Data stored permanently
Easy data management
Useful for projects
Helps store user information
