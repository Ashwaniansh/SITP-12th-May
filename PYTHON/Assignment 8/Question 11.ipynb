```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv(r"D:\Data Science with AI & ML\SITP-12th-May\PANDAS\sales.csv")
sns.set_style("whitegrid")
fig = plt.figure(figsize=(18, 5))
# Scatter Plot
ax1 = plt.subplot(1, 3, 1)
sns.scatterplot(
    data=df,
    x="Units Sold",
    y="Total Revenue",
    hue="Item Type",
    ax=ax1
)
ax1.set_title("Scatter: Units Sold vs Revenue")
# Pair Plot
plt.tight_layout()
plt.show()
sns.pairplot(
    df[["Units Sold", "Unit Price", "Total Revenue", "Total Profit"]],
    diag_kind="hist"
)
# Joint Plot 
sns.jointplot(
    data=df,
    x="Units Sold",
    y="Total Revenue",
    kind="scatter",
    height=6
)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
np.random.seed(42)
df = pd.DataFrame({
    "feature1": np.random.normal(50, 10, 100),
    "feature2": np.random.normal(30, 5, 100)
})
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x="feature1", y="feature2")
plt.title("Scatter Plot: Feature1 vs Feature2")
plt.show()
sns.pairplot(df)
plt.suptitle("Pair Plot of Features", y=1.02)
plt.show()
sns.jointplot(data=df, x="feature1", y="feature2", kind="scatter")
plt.suptitle("Joint Plot: Feature1 vs Feature2", y=1.02)
plt.show()
```
