```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1, 11)
y = np.array([12, 18, 25, 30, 22, 35, 40, 38, 45, 50])
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
# Line Plot
axes[0, 0].plot(x, y, marker='o')
axes[0, 0].set_title("Line Plot")
axes[0, 0].set_xlabel("X values")
axes[0, 0].set_ylabel("Y values")
# Bar Plot 
axes[0, 1].bar(x, y)
axes[0, 1].set_title("Bar Plot")
axes[0, 1].set_xlabel("X values")
axes[0, 1].set_ylabel("Y values")
# Scatter Plot
axes[1, 0].scatter(x, y)
axes[1, 0].set_title("Scatter Plot")
axes[1, 0].set_xlabel("X values")
axes[1, 0].set_ylabel("Y values")
# Histogram 
axes[1, 1].hist(y, bins=5, edgecolor='black')
axes[1, 1].set_title("Histogram")
axes[1, 1].set_xlabel("Value Range")
axes[1, 1].set_ylabel("Frequency")
plt.tight_layout()
plt.show()
```
