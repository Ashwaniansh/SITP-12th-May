```python
import pandas as pd
df = pd.read_csv(r"D:\Data Science with AI & ML\SITP-12th-May\PANDAS\loan_approval.csv")
missing_percent = df.isnull().mean() * 100
print("Missing Percentage in Each Column:")
print(missing_percent)
cols_to_remove = missing_percent[missing_percent > 20].index
print("\nColumns to Remove (>20% missing values):")
print(list(cols_to_remove))
df_cleaned = df.drop(columns=cols_to_remove)
print("\nRemaining Columns:")
print(df_cleaned.columns.tolist())
```
