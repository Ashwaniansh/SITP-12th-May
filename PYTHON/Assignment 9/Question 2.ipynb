```python
class Calculator:
  def add(self, a, b):
    return a + b
  def sub(self, a, b):
    return a - b
  def mul(self, a, b):
    return a * b
  def div(self, a, b):
    if b != 0:
        return a / b
    else:
        return "Cannot divide by zero"
cal = Calculator()
print("Addition:", cal.add(10, 5))
print("Subtraction:", cal.sub(10, 5))
print("Multiplication:", cal.mul(10, 5))
print("Division:", cal.div(10, 5))
```
