```python
class Employee:
  def __init__(self, emp_name, emp_id, salary):
      self.emp_name = emp_name
      self.emp_id = emp_id
      self.salary = salary
  def display(self):
      print("Employee Name:", self.emp_name)
      print("Employee ID:", self.emp_id)
      print("Salary:", self.salary)
      print("----------------------")
e1 = Employee("Ashwani Kumar", 101, 50000)
e2 = Employee("Anshu", 102, 60000)
e3 = Employee("Arnav", 103, 55000)
e1.display()
e2.display()
e3.display()
```
