* Inheritance is a feature of OOP (Object-Oriented Programming) where one class can use the properties and methods of another class.

* Parent Class / Base Class → The class whose properties are inherited.
* Child Class / Derived Class → The class that inherits properties.

        Inheritance helps in:

* Code reusability
* Reducing duplicate code
* Easy maintenance


Types of Inheritance in Python


| Type                     | Meaning                                    |
| ------------------------ | ------------------------------------------ |
| Single Inheritance       | One parent → One child                     |
| Multiple Inheritance     | Multiple parents → One child               |
| Multilevel Inheritance   | Grandparent → Parent → Child               |
| Hierarchical Inheritance | One parent → Multiple children             |
| Hybrid Inheritance       | Combination of different inheritance types |

```python
# Single Inheritance

class College:
  pass
class Student(College):
  pass
```

```python
class College:
  college_name = "SRMIST"
  college_address = "GHAZIABAD"
  collegesp = 50
  def show(self):
    print(self.college_name)
    print(self.college_address)
    print(self.collegesp)

class Student(College):
   student_name = "Ashwani"
   student_rollnumber = 2
   student_number = 9102376887
   student_branch = "MCA"
st = Student()
st.show()
print(st.student_branch)
```

```python
# Multilevel Inheritance
class Person:
  def info(self):
    print("I am Person")
class Teacher(Person):
  def teacher_info(self):
    print("I am a Teacher")
class HOD(Teacher):
  def hod_info(self):
    print("I am HOD")
h = HOD()
print(h.hod_info())          # inherited from Person
print(h.teacher_info())  # inherited from Teacher
print(h.info())
```

```python
class Employees():
  def Name(self):
    print("Employee Name: Ashwani")
class salary(Employees):
  def Salary(self):
    print("Salary: 10000")
class Designation(salary):
   def desig(self):
     print ("Designation: Test Engineer")
call = Designation()
call.Name()
call.Salary()
call.desig()
```

```python
# Mult[]
class Calculation1:
   def Sum(self,a,b):
     return a+b;
class Calculation2:
   def Multi(self,a,b):
     return a*b;
class Derived(Calculation1,Calculation2):
   def Divide(self,a,b):
     return a/b;
d = Derived()
print(d.Sum(10,20))
print(d.Multi(10,20))
print(d.Divide(10,20))
```

```python
class Company1:
    def company1_info(self):
        print("Company 1: TATA")
class Company2:
    def company2_info(self):
        print("Company 2: MAHNDRA")
class Company3(Company1, Company2): 
    def company3_info(self):
        print("Company 3: Accessing data from Company1 and Company2")
obj = Company3()
obj.company1_info()
obj.company2_info()
obj.company3_info()
```

```python
# Multiple Inheritance
class Developer:
    def coding(self):
        print("Code Written")
class Manager:
    def meeting(self):
        print("Code Checking")
class Teamleader(Developer, Manager):
    def project(self):
        print("Project Handle")
team = Teamleader()

team.coding()
team.meeting()
team.project()
```

```python
# Hierarchical Inheritance
class Developer:
    def coding(self):
        print("Code Written")
class Manager:
    def meeting(self):
        print("Code Checking")
class Teamleader(Developer, Manager):
    def project(self):
        print("Project Handle")

class Designer(Developer):
    print("Designer")

team = Teamleader()
man = Manager()
des = Designer()
team.coding()
```

```python
# Hybrid Inheritance
class Person:
    def display(self):
        print("Person Details.")
class Student(Person):
    def study(self):
        print("Student is studying.")
class Teacher(Person):
    def teach(self):
        print("Teacher is teaching.")
class School(Student, Teacher):
    def school_info(self):
        print("School Information.")
obj = School()
obj.display()
obj.study()
obj.teach()
obj.school_info()
```
