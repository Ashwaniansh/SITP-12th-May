#### python Function 
* A function is a block of reusable code that performs a specific task.
Functions help make code shorter, organized, and reusable.


Types of Functions in Python

There are mainly 2 types of functions in Python:

1. Built-in Functions

These functions are already available in Python.
Some Common Built-in Functions
* print()
* len()
* max()
* min()
* sum()
* type()
* input()

2. User-defined Functions

These functions are created by the user using the **def** keyword.

```python
# 1. Parameter

# A parameter is a variable written inside the function definition.

# Example
# def greet(name):
#     print("Hello", name)

# Here:

# name is a parameter
# 2. Argument

# An argument is the actual value passed to the function when calling it.

# Example
# def greet(name):
#     print("Hello", name)

# greet("Rahul")
# Output
# Hello Rahul

# Here:

# "Rahul" is an argument
# Simple Difference
# Parameter	Argument
# Variable in function definition	Actual value passed
# Receives value	Sends value
# Example with Two Parameters
# def add(a, b):
#     print(a + b)

# add(10, 20)
# Output
# 30
# Explanation
# a and b → parameters
# 10 and 20 → arguments
# Types of Arguments
# 1. Positional Arguments
# def student(name, age):
#     print(name, age)

# student("Aman", 20)
# 2. Keyword Arguments
# def student(name, age):
#     print(name, age)

# student(age=20, name="Aman")
# 3. Default Arguments
# def country(name="India"):
#     print(name)

# country()
# Output
# India
```

```python
def check(num):
    if num % 2 == 0:
        print(num, "This is even No")
    else:
        print(num, "This is odd number")

check(10)
check(7)
```

```python
def check(numbers):
    for i in numbers:
        if i % 2 == 0:
            print(i, "Even")
        else:
            print(i, "Odd")
lst = [10, 15, 20, 25]
check(lst)
```

```python
def even_odd(lst):
    for x in lst:
        if x % 2 == 0:
            print(x, "This is even number")
        else:
            print(x, "This is odd number")
lst1 = [1, 2, 3, 4, 5, 6, 7, 8]
even_odd(lst1)
```

```python
def prime(n):
    if n > 1:
        for i in range(2, n):
            if n % i == 0:
                print(n, "Not Prime")
                break
        else:
            print(n, "Prime")
    else:
        print(n, "Not Prime")
x = int(input("Enter number: "))
prime(x)
```
