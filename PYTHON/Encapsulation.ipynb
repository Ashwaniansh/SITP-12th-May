Encapsulation in Python

Encapsulation is an OOP (Object-Oriented Programming) concept used to wrap data (variables) and methods (functions) together inside a class and restrict direct access to data for protection.

It helps in data hiding and allows controlled access using methods.

Encapsulation = Data + Methods + Access Control
Encapsulation Explained in a Table
Term	Meaning	Example
Class	Blueprint to create objects	BankAccount
Data (Variable)	Stores information	__balance
Method (Function)	Performs operations on data	deposit()
Encapsulation	Combining data and methods in one class	BankAccount class
Data Hiding	Prevents direct access to data	__balance
Controlled Access	Access data through methods	show_balance()
Example: Bank Account Using Encapsulation
class BankAccount:

    def __init__(self):
        self.__balance = 0      # Private variable

    def deposit(self, amount):
        self.__balance += amount
        print("Deposited:", amount)

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient Balance")

    def show_balance(self):
        print("Current Balance:", self.__balance)


# Create object
acc = BankAccount()

acc.deposit(5000)
acc.withdraw(2000)
acc.show_balance()
Code Explanation (Step by Step)
Code	Explanation
class BankAccount:	Creates a class
__init__()	Constructor runs automatically
self.__balance = 0	Private variable (__) hides data
deposit()	Adds money
withdraw()	Removes money if balance exists
show_balance()	Displays balance
acc = BankAccount()	Creates object
acc.deposit(5000)	Deposits money
acc.withdraw(2000)	Withdraws money
acc.show_balance()	Shows remaining balance
Output
Deposited: 5000
Withdrawn: 2000
Current Balance: 3000
Why Encapsulation is Used
Benefit	Description
Security	Protects data from direct changes
Data Hiding	Keeps variables private
Easy Maintenance	Changes can be managed inside class
Controlled Access	Data accessed through methods only
Example of restricted access:
acc.__balance = 10000

This will not directly change the private balance because __balance is encapsulated inside the class.
