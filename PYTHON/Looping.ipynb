```python
# LOOP
# A loop is used to execute a block of code repeatedly until a condition becomes false or the sequence ends.

# Loops help reduce repeated code and save time.

# Types of Loops in Python

# Python has mainly 2 types of loops:

# for loop
# while loop
# 1. for Loop

# A for loop is used when the number of iterations is known.

# Syntax
# for variable in sequence:
#     # code
# Example
# for i in range(1, 6):
#     print(i)
# Output
# 1
# 2
# 3
# 4
# 5
# 2. while Loop

# A while loop is used when the condition needs to be checked repeatedly.

# Syntax
# while condition:
#     # code
# Example
# x = 1

# while x <= 5:
#     print(x)
#     x += 1
# Output
# 1
# 2
# 3
# 4
# 5
# Difference Between for and while Loop
# for Loop	                      while Loop
# Used for fixed iterations	      Used for condition-based iterations
# Easier to use with sequences	  Runs until condition becomes False
# Loop Control Statements
# Statement	Purpose
# break	    Stops the loop
# continue	Skips one iteration
# pass	    Does nothing
```
