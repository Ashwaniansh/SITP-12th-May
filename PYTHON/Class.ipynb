Class Definition in Python

A class is a blueprint or template used to create objects.

It contains:

variables (data)
methods (functions)

Classes are used in OOPS (Object-Oriented Programming System).

Syntax
class ClassName:
    # variables
    # methods
Example
class Student:

    name = "Rahul"

    def show(self):
        print("Hello")


s1 = Student()

print(s1.name)
s1.show()
Output
Rahul
Hello
Explanation
Part	Meaning
class Student	Creates class
name	Variable
show()	Method
s1	Object
Pillars of OOPS

There are 4 main pillars of OOPS.

Pillar	Meaning
Inheritance	Acquiring properties from another class
Polymorphism	One function, many forms
Encapsulation	Data hiding
Abstraction	Hiding implementation details
1. Inheritance

Child class uses parent class properties.

class Father:

    def house(self):
        print("House")


class Son(Father):
    pass


s1 = Son()

s1.house()
Inheritance Formula

Child Class→Parent Class

2. Polymorphism

Same method behaves differently.

print(len("Python"))
print(len([1, 2, 3]))
3. Encapsulation

Wrapping data and methods together.

class Bank:

    def __init__(self):
        self.__balance = 1000
4. Abstraction

Hides internal details and shows only necessary features.

Example:

ATM machine
Car driving system
