Constructor in Python

A constructor is a special method that runs automatically when an object is created.

In Python, the constructor method is:

__init__()

It is mainly used to initialize object variables (attributes).

Constructor Explained in a Table
Term	Meaning
Constructor	Special method called automatically
Method Name	__init__()
Purpose	Initialize object data
Executes	When object is created
First Parameter	self
Example 1: Constructor with Values
class Student:

    def __init__(self):
        print("Constructor Called")

obj = Student()
Output:
Constructor Called
Explanation:
obj = Student() creates an object.
Python automatically calls __init__().
Example 2: Constructor to Store Data
class Student:

    def __init__(self, name):
        self.name = name

    def show(self):
        print("Name:", self.name)

obj = Student("Ansh")

obj.show()
Output:
Name: Ansh
Code Explanation
Code	Meaning
__init__()	Constructor
self.name	Instance variable
name	Value passed
obj = Student("Ansh")	Creates object
obj.show()	Displays stored value
Flow
obj = Student("Ansh")
        ↓
__init__() runs automatically
        ↓
self.name = "Ansh"
        ↓
show()
        ↓
Name: Ansh

```python
#Multilevel Inheritance using constructor
class Vehicle:
    def __init__(self, company):
        self.company = company
class Car(Vehicle):
    def __init__(self, company, model):
        Vehicle.__init__(self, company)
        self.model = model
class SpecialCar(Car):
    def __init__(self, company, model, speed):
        Car.__init__(self, company, model)
        self.speed = speed
    def display(self):
        print("Company Name:", self.company)
        print("Model type:", self.model)
        print("Top Speed:", self.speed)
obj = SpecialCar("Mahindra", "XUV7XO", "220km/h")
obj.display()
```

```python
# Multiple Inheritance using constructor
class Student:
    def __init__(self, name):
        self.name = name
class Course:
    def __init__(self, branch):
        self.branch = branch
class College(Student, Course):
    def __init__(self, name, branch):
        Student.__init__(self, name)
        Course.__init__(self, branch)
    def display(self):
        print("Name:", self.name)
        print("Branch:", self.branch)
obj = College("Ashwani", "MCA")
obj.display()
```
