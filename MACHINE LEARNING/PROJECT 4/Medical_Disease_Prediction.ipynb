```python
df = df.drop_duplicates()
```

```python

for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype(str).str.strip()
df.head()
```

```python

for col in df.select_dtypes(include='object').columns:
    print(col, df[col].unique()[:10])
```

```python

numeric_cols = df.select_dtypes(include='number').columns
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    print(col, ((df[col] < lower) | (df[col] > upper)).sum())
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.heatmap(
    df.isnull(),
    cmap="YlGnBu",      
    cbar=True,          
    yticklabels=False,
    linewidths=0.1,
    linecolor="white"
)
plt.title("Missing Values Heatmap", fontsize=14, fontweight="bold")
plt.xlabel("Features")
plt.ylabel("Records")
plt.tight_layout()
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')
plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
numeric_df.hist(figsize=(15,10))
plt.tight_layout()
plt.show()
```

```python
for col in numeric_df.columns[:6]:
  plt.figure(figsize=(6,3))
  sns.boxplot(x=df[col])
  plt.title(f"Boxplot of {col}")
  plt.show()
```

```python
target_col = df.columns[-1]
plt.figure(figsize=(8,5))
sns.countplot(y=df[target_col], order=df[target_col].value_counts().index)
plt.title("Disease Distribution")
plt.show()
```

```python
cols = list(numeric_df.columns[:4])
if len(cols) >= 2:
    sns.pairplot(df[cols])
    plt.show()
```

```python
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols[:3]:
  plt.figure(figsize=(8,4))
  sns.countplot(y=df[col], order=df[col].value_counts().head(10).index)
  plt.title(f"Distribution of {col}")
  plt.show()
```
