Machine Learning (ML) is a branch of Artificial Intelligence that enables computers to learn from data and improve their performance without being explicitly programmed for every task.

### Machine learning 
* Machine Learning is a method of teaching computers to learn from data and improve their performance automatically.


| Type                     | Description                          |
| ------------------------ | ------------------------------------ |
| Supervised Learning      | Learns from labeled data             |
| Unsupervised Learning    | Learns from unlabeled data           |
| Semi-Supervised Learning | Uses both labeled and unlabeled data |
| Reinforcement Learning   | Learns through rewards and penalties |

| Labeled Data                 | Unlabeled Data                   |
| ---------------------------- | -------------------------------- |
| Has correct answers (labels) | No correct answers               |
| Used in Supervised Learning  | Used in Unsupervised Learning    |
| Easier to train models       | Used to discover hidden patterns |
| Example: House Price Dataset | Example: Customer Purchase Data  |
