```python
df.columns = [c.strip().lower().replace(' ','_') for c in df.columns]
df.columns
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
```

```python
df.isnull().sum()
```

```python
print('Duplicates:', df.duplicated().sum())
df = df.drop_duplicates()
```

```python
df.describe(include='all')
```

```python
numeric_df = df.select_dtypes(include=np.number)
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
profit_region = df.groupby('region')['total_profit'].sum().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=profit_region.index, y=profit_region.values)
plt.xticks(rotation=45)
plt.title('Total Profit by Region')
plt.show()
```

```python
top_country = df.groupby('country')['total_revenue'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_country.values, y=top_country.index)
plt.title('Top 10 Countries by Revenue')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, y='item_type', order=df['item_type'].value_counts().index)
plt.title('Item Type Distribution')
plt.show()
```

```python
sns.histplot(df['units_sold'], kde=True)
plt.title('Distribution of Units Sold')
plt.show()
```

```python
sns.scatterplot(data=df, x='total_revenue', y='total_profit', hue='sales_channel')
plt.title('Revenue vs Profit')
plt.show()
```

```python
sns.boxplot(data=df, x='sales_channel', y='total_profit')
plt.title('Profit by Sales Channel')
plt.show()
```

```python
monthly = df.groupby(df['order_date'].dt.to_period('M'))['total_revenue'].sum()
monthly.index = monthly.index.astype(str)
plt.figure(figsize=(14,5))
sns.lineplot(x=monthly.index, y=monthly.values, marker='o')
plt.xticks(rotation=90)
plt.title('Monthly Revenue Trend')
plt.show()
```

```python
sns.countplot(data=df, x='order_priority')
plt.title('Order Priority Distribution')
plt.show()
```
