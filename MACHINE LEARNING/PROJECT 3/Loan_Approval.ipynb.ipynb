```python

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(12,5))
sns.heatmap(df.isnull(), cmap='YlGnBu', cbar=True, yticklabels=False)
plt.title('Missing Values Heatmap')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
target = df.columns[-1]
plt.figure(figsize=(7,5))
sns.countplot(x=df[target])
plt.title('Loan Approval Distribution')
plt.show()
```

```python
numeric_df.hist(figsize=(15,10))
plt.tight_layout()
plt.show()
```

```python
for col in numeric_df.columns[:6]:
  plt.figure(figsize=(8,4))
  sns.boxplot(x=df[col])
  plt.title(f'Boxplot of {col}')
  plt.show()
```

```python
cols = list(numeric_df.columns[:4])
if len(cols) >= 2:
  sns.pairplot(df[cols])
  plt.show()
```

```python
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols[:4]:
  plt.figure(figsize=(8,4))
  sns.countplot(y=df[col], order=df[col].value_counts().head(10).index)
  plt.title(f'Distribution of {col}')
  plt.show()
```
