Key Features of CART

Classification Tree: Predicts categorical outcomes (e.g., Yes/No, Spam/Not Spam).
Regression Tree: Predicts continuous values (e.g., house price, temperature).
Binary Splits

Model Evaluation Techniques in Machine Learning

Model evaluation techniques are used to measure how well a machine learning model performs on unseen data.

1. Train-Test Split
Dataset is divided into two parts:
Training Set (e.g., 80%)
Testing Set (e.g., 20%)
Model is trained on the training set and evaluated on the test set.

Advantages:

Simple and fast.

Disadvantages:

Performance depends on how data is split.
2. Cross-Validation
k-Fold Cross-Validation
Dataset is divided into k equal folds.
Model is trained on k−1 folds and tested on the remaining fold.
Process repeats k times.
Final score = Average of all scores.

Example (5-Fold):

Fold1 → Test, Fold2-5 → Train
Fold2 → Test, Fold1,3,4,5 → Train
...

Advantages:

More reliable than train-test split.
Uses all data for training and testing.
3. Leave-One-Out Cross-Validation (LOOCV)

Special case of k-fold where:

k=n

(n = number of observations)

One sample is used for testing and the rest for training.

Advantage: Uses maximum data for training.

Disadvantage: Computationally expensive.

4. Stratified Cross-Validation
Maintains the same class distribution in each fold as in the original dataset.
Useful for imbalanced datasets.
5. Bootstrap Method
Random sampling with replacement.
Some samples may appear multiple times.
Used when dataset is small.
