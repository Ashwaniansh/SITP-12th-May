```python
a = [1,2,3,4]
plt.pie(a)
plt.show()
```

```python
a = [1,2,3,4]
label = ["Apple", "Mango", "Lichi", "Banana"]
plt.pie(a, labels=label)
plt.show()
```

```python
a = [1,2,3,4]
label = ["Apple", "Mango", "Lichi", "Banana"]
plt.pie(a, labels=label, explode = [0,0.2,0,0])
plt.show()
```

```python
a = [1,2,3,4]
label = ["Apple", "Mango", "Lichi", "Banana"]
plt.pie(a, labels=label, explode = [0,0.2,0,0], shadow=True)
plt.show()
```

```python
a = [1,2,3,4]
label = ["Apple", "Mango", "Lichi", "Banana"]
colour = ['pink','green','yellow','blue']
plt.pie(a, labels=label, explode = [0,0.2,0,0], shadow=True, colors = colour)
plt.show()
```

```python
a = [1,2,3,4]
label = ["Apple", "Mango", "Lichi", "Banana"]
plt.pie(a, labels=label, explode = [0,0.2,0,0], shadow=True, autopct='%1.1f%%')
plt.show()
```

```python
a = [1,2,3,4]
label = ["Apple", "Mango", "Lichi", "Banana"]
colour = ['pink','green','yellow','blue']
plt.pie(a, labels=label, explode = [0,0.2,0,0], shadow=True, autopct='%1.1f%%', colors = colour)
plt.legend()
plt.show()
```
